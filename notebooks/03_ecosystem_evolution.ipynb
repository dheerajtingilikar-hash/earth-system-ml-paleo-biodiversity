```python
import pandas as pd

df = pd.read_csv(r"F:\EARTH_SYSTEM_ML_PROJECT\data\raw\pbdb\pbdb_fossils.csv",
                 low_memory=False)

df.head()
```

```python
df = df.dropna(subset=["early_interval", "accepted_name"])
```

```python
diversity = df.groupby("early_interval")["accepted_name"].nunique()

diversity = diversity.sort_index()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(diversity.index, diversity.values, marker="o")
plt.xticks(rotation=90)
plt.title("Biodiversity Through Time (PBDB)")
plt.ylabel("Number of Genera")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(diversity.index, diversity.values, marker="o")

plt.axvline(x="Changhsingian", color="red", linestyle="--", label="P–T Extinction")

plt.xticks(rotation=90)
plt.title("Mass Extinction Signal in Biodiversity")
plt.legend()
plt.show()
```

```python
occ = df.groupby("early_interval")["occurrence_no"].count()

plt.figure(figsize=(12,5))
plt.bar(occ.index, occ.values)
plt.xticks(rotation=90)
plt.title("Fossil Occurrence Through Time")
plt.show()
```
